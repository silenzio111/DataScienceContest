```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

pwd = '../data/test.xlsx'
output_path = '../data(processed)/test.csv'
```

```python
df = pd.read_excel(pwd)
df.head()
```

```python
df.info()
```

```python
# 查看每列有多少唯一值
df.nunique()
```

```python
df.describe()
```

```python
# 找出非数值型的列
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns
non_numeric_cols
```

```python
df[non_numeric_cols].nunique()
```

```python
# laber encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

mapping_dict = {}
for col in non_numeric_cols:
    df[col] = le.fit_transform(df[col])
    # 输出mapping dict，对应原始值和编码值
    mapping_dict[col] = dict(zip(le.classes_, le.transform(le.classes_)))
print(mapping_dict)
```

```python

df.to_csv(output_path, index=False)
```
