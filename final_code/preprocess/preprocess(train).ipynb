```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

pwd = '../data/train.xlsx'
output_path = '../data(processed)/train.csv'
```

```python
df = pd.read_excel(pwd)
df.head()
```

```python
# 找出非数值型的列
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns
non_numeric_cols
```

```python
df[non_numeric_cols].nunique()
```

```python
mapping_dict = {
    'housing': {'其他': np.int64(0), '租赁': np.int64(1),
                '自有无贷': np.int64(2), '自有有贷': np.int64(3)},
    'purpose': {'个人综合消费贷': np.int64(0), '信用卡账单分期': np.int64(1),
                '债务优化重组': np.int64(2), '医疗支出分期': np.int64(3),
                '大件消费贷': np.int64(4), '婚庆分期': np.int64(5),
                '家装分期': np.int64(6), '小微经营贷': np.int64(7),
                '旅游消费贷': np.int64(8), '汽车消费贷': np.int64(9),
                '租房押金贷': np.int64(10), '绿色家电分期': np.int64(11),
                '购房贷': np.int64(12)}
}

# 对应mapping_dict进行映射
for col, mapping in mapping_dict.items():
    df[col] = df[col].map(mapping)
df.head()




```
